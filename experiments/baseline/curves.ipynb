```python
import pandas as pd
import matplotlib.pyplot as plt

# load history.csv
df = pd.read_csv("history.csv")

# columns are epoch	train_loss	valid_loss	multi_label_accuracy	multilabel_hamming_loss	macro_f1_score_multilabel
```

```python
# plot train and valid loss vs epoch
plt.plot(df.epoch, df.train_loss, label='train_loss')
plt.plot(df.epoch, df.valid_loss, label='valid_loss')

plt.legend()
# plt.xlim(0, 5)
# plt.ylim(0.1, 0.3)
```

```python
import numpy as np

# print row with the best validation loss
row_indx = np.argmin(df["valid_loss"])
df.iloc[row_indx]
```
