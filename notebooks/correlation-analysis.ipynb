```python

import pandas as pd
import numpy as np
import wfdb
import ast
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '../data/physionet.org/files/ptb-xl/1.0.2/'
sampling_rate=100
```

# Load data

```python
def load_raw_data(df, sampling_rate, path):
    if sampling_rate == 100:
        data = [wfdb.rdsamp(path+f) for f in df.filename_lr]
    else:
        data = [wfdb.rdsamp(path+f) for f in df.filename_hr]
    data = np.array([signal for signal, meta in data])
    return data
```

```python

# load and convert annotation data
Y = pd.read_csv(path+'ptbxl_database.csv', index_col='ecg_id')
Y.scp_codes = Y.scp_codes.apply(lambda x: ast.literal_eval(x))


# Load scp_statements.csv for diagnostic aggregation
agg_df = pd.read_csv(path+'scp_statements.csv', index_col=0)
agg_df = agg_df[agg_df.diagnostic == 1]

def aggregate_diagnostic(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_class)
    return list(set(tmp))

def aggregate_diagnostic_subclass(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_subclass)
    return list(set(tmp))

# Apply diagnostic superclass
Y['diagnostic_superclass'] = Y.scp_codes.apply(aggregate_diagnostic)
Y['diagnostic_subclass'] = Y.scp_codes.apply(aggregate_diagnostic_subclass)


Y
```

```python

# Load raw signal data
X = load_raw_data(Y, sampling_rate, path)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm

X_data = X.copy()  # shape (21801, 1000, 12)

# Randomly select 1024 samples
num_samples = 1024
selected_samples = np.random.choice(X_data.shape[0], num_samples, replace=False)
sampled_data = X_data[selected_samples]

# Compute and accumulate the correlation matrix for each sample
cumulative_corr_matrix = np.zeros((12, 12))
for sample in tqdm(sampled_data, total=len(sampled_data)):
    df = pd.DataFrame(sample, columns=[f'Feature {i+1}' for i in range(12)])
    cumulative_corr_matrix += df.corr().values

# Average the correlation matrices
average_corr_matrix = cumulative_corr_matrix / num_samples

# Visualize the average correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(average_corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=.5)
plt.show()
```
