```python

import pandas as pd
import numpy as np
import wfdb
import ast
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '../data/physionet.org/files/ptb-xl/1.0.2/'
sampling_rate=100
```

# Load data

```python
def load_raw_data(df, sampling_rate, path):
    if sampling_rate == 100:
        data = [wfdb.rdsamp(path+f) for f in df.filename_lr]
    else:
        data = [wfdb.rdsamp(path+f) for f in df.filename_hr]
    data = np.array([signal for signal, meta in data])
    return data
```

```python

# load and convert annotation data
Y = pd.read_csv(path+'ptbxl_database.csv', index_col='ecg_id')
Y.scp_codes = Y.scp_codes.apply(lambda x: ast.literal_eval(x))


# Load scp_statements.csv for diagnostic aggregation
agg_df = pd.read_csv(path+'scp_statements.csv', index_col=0)
agg_df = agg_df[agg_df.diagnostic == 1]

def aggregate_diagnostic(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_class)
    return list(set(tmp))

def aggregate_diagnostic_subclass(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_subclass)
    return list(set(tmp))

# Apply diagnostic superclass
Y['diagnostic_superclass'] = Y.scp_codes.apply(aggregate_diagnostic)
Y['diagnostic_subclass'] = Y.scp_codes.apply(aggregate_diagnostic_subclass)


Y
```

```python

# Load raw signal data
X = load_raw_data(Y, sampling_rate, path)
```

```python
X.shape
```

```python
import numpy as np

x = X[0, :, 0]
x = x - np.mean(x)
x_fft = np.fft.fftn(x)
x_fft_smooth = x_fft.copy()
x_fft_smooth[100:900] = 0
x_smooth_ifft = np.fft.ifftn(x_fft_smooth)

plt.plot(x_fft[:300])
plt.plot(x_fft_smooth[:300])
```

```python
plt.plot(x[200:400], label='original')
plt.plot(x_smooth_ifft[200:400], label='smoothed')
```

```python
plt.plot(x, label='original')
plt.plot(x_smooth_ifft, label='smoothed')
```

Interesting how you can use so little fourier components to get a good approximation of the original signal.
